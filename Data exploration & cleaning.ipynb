### remove student & affordable housing

```python
remove_cols = ['PROPERTY_STUDENTHOUSING','PROPERTY_AFFORDABLEHOUSING', 'PROPERTY_MILITARYHOUSING','PROPERTY_AGERESTRICTED', 'PROPERTY_FRACTUREDCONDO']

clean_data = data

for i in remove_cols:
  clean_data = clean_data.drop(clean_data.loc[(clean_data[i] == 'A')|(clean_data[i] == 'P')].index)

clean_data.head().transpose()
```

```python
# original
data['PROPERTY_CURRENT_RENT'].isna().sum()
```

```python
# new
clean_data['PROPERTY_CURRENT_RENT'].isna().sum()
```

```python
# make sure correct rows were deleted

columns = ['PROPERTY_MIXEDUSE', 'PROPERTY_STUDENTHOUSING',
       'PROPERTY_AFFORDABLEHOUSING', 'PROPERTY_MILITARYHOUSING',
       'PROPERTY_AGERESTRICTED', 'PROPERTY_FRACTUREDCONDO',
       'PROPERTY_ADAPTIVEREUSE', 'PROPERTY_SINGLEFAMILYRENTAL']

for col in columns:
  print(col)
  print(clean_data[col].unique())
```

```python
# remove useless columns

new_remove_cols = ['PROPERTY_MIXEDUSE', 'PROPERTY_MIXEDUSE_NOTES','PROPERTY_STUDENTHOUSING',
       'PROPERTY_AFFORDABLEHOUSING', 'PROPERTY_MILITARYHOUSING',
       'PROPERTY_AGERESTRICTED', 'PROPERTY_FRACTUREDCONDO',
       'PROPERTY_ADAPTIVEREUSE', 'PROPERTY_SINGLEFAMILYRENTAL']

clean_data = clean_data.drop(columns = new_remove_cols)
```

```python
# remove all data points where current rent is missing

clean_data = clean_data.dropna(subset = ['PROPERTY_CURRENT_RENT'])
```
